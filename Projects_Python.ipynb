```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df=pd.read_csv("Financial.csv",sep=",")
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated()
```

```python
df.nunique()
```

```python
df.columns
```

```python
df['Country'].value_counts()
```

```python
Profit_per_country = df.groupby('Country')[' Profit '].sum()
Profit_per_country
```

```python
plt.bar(Profit_per_country.index, Profit_per_country.index)
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.title('Total Profit by Country')
plt.show()
```

```python
sales_per_country = df.groupby('Country')['  Sales '].sum()

plt.bar(sales_per_country.index, sales_per_country.values)
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.title('Total Sales by Country')
plt.show()
```

```python
profit_per_year = df.groupby('Year')[' Profit '].sum()

plt.plot(profit_per_year.index, profit_per_year.index, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.title('Total Profit per Year')
plt.show()
```

```python
sns.scatterplot(x='Units Sold', y=' Profit ', data=df)
plt.xlabel('Units Sold')
plt.ylabel('Profit')
plt.title('Scatter plot: Units Sold vs. Profit')
plt.show()
```

```python
numerical_columns = df.select_dtypes(include='number')
numerical_columns 
```

```python
correlation_matrix = numerical_columns.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Correlation Plot')
plt.show()
```

```python
selected_columns = ['Units Sold',' Manufacturing Price ', ' Sale Price ', ' Gross Sales ', ' Discounts ', ' Profit ']

sns.pairplot(df[selected_columns])
plt.suptitle('Pair Plot of Selected Numerical Columns', y=1.02)
plt.show()
```

```python
profit_by_segment = df.groupby('Segment')[' Profit '].sum()

profit_by_country = df.groupby('Country')[' Profit '].sum()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(profit_by_segment.index, profit_by_segment.index)
plt.xlabel('Segment')
plt.ylabel('Total Profit')
plt.title('Total Profit by Segment')


plt.subplot(1, 2, 2)
plt.bar(profit_by_country.index, profit_by_country.index)
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.title('Total Profit by Country')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
sales_data = df['  Sales ']

plt.figure(figsize=(8, 6))
plt.hist(sales_data, bins=20, edgecolor='black')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Histogram of Sales')
plt.grid()
plt.show()
```

```python
sales_per_country = df.groupby('Country')['  Sales '].sum()

plt.figure(figsize=(8, 8))
plt.pie(sales_per_country, labels=sales_per_country.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Sales per Country')
plt.show()
```
